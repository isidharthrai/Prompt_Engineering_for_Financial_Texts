```python
#load data from ../../DatasetAnalysis_FinancialPhraseBank/processed_100pct_agreement.csv
import pandas as pd

df = pd.read_csv('../../DatasetAnalysis_FinancialPhraseBank/processed_100pct_agreement.csv')

print("Columns:", df.columns.tolist())
print("Value counts for 'sentiment':")
print(df['sentiment'].value_counts())
```

```python
import pandas as pd

# 1. Sort the entire DataFrame by length metrics in descending order
df_sorted = df.sort_values(by=['char_length', 'word_count'], ascending=False)

# 2. Group by 'sentiment' and extract the top 70 from each category
processed_df = df_sorted.groupby('sentiment').head(70)

# 4. Final selection of columns and resetting the index for a clean output
processed_df = processed_df[['sentence', 'sentiment','word_count','char_length']].reset_index(drop=True)

# Verification: Check the counts for each sentiment
print("Value counts for 'sentiment' in the processed DataFrame:")
print(processed_df['sentiment'].value_counts())
```

```python
print("Processed DataFrame:")
print(processed_df.head())
```

```python
# This will show you the highest word count for each sentiment group
print("Max word counts in your processed data:")
print(processed_df.groupby('sentiment')['word_count'].max())

# This will show you the shortest word count in each group (the '70th' row)
print("\nMinimum word counts (the 70th row) in your processed data:")
print(processed_df.groupby('sentiment')['word_count'].min())
```

```python
#Save the processed DataFrame to a new CSV file
processed_df.to_csv('processed_data.csv', index=False)
```
